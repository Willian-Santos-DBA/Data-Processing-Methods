```python
df[pd.isnull(df['body_type'])]
```

```python
# Filtra os dados removendo os nulos apenas para construir o dicionário confiável
```

```python
df_valid_body = df.dropna(subset=['body_type'])
```

```python
# Agrupa por modelo e body_type
```

```python
df_grouped_model_body = df_valid_body.groupby(['model', 'body_type']).size().reset_index()
```

```python
df_grouped_model_body
```

```python
# Cria o dicionário iterando sobre o agrupamento
```

```python
model_body_dict = {}
for index, row in df_grouped_model_body.iterrows():
  model_body_dict[row['model']] = row['body_type']
```

```python
def fill_body_type(body_type, model, model_body_dict):
  if pd.isnull(body_type):
    return model_body_dict.get(model, body_type)
  else:
    return body_type
```

```python
df[['body_type', 'model']].apply(lambda x: fill_body_type(x['body_type'], x['model'], model_body_dict), axis=1).iloc[[2, 31, 63, 146, 187]]
```

```python
df['body_type'] = df.apply(lambda x : fill_body_type(x['body_type'], x['model'], model_body_dict), axis=1)
```

```python
df.groupby(['brand', 'model', 'body_type', 'market_segment']).agg({'annual_sales_units':'mean'}).reset_index()
```

```python
# O transform calcula as médias baseado no grupo, mas mantém o tamanho exato da tabela original
```

```python
medias_calculadas = df.groupby(['brand', 'model', 'body_type', 'market_segment'])['annual_sales_units'].transform('mean')
```

```python
# Agora pedimos para o Pandas preencher os nulos da coluna original usando essas médias
```

```python
df['annual_sales_units'] = df['annual_sales_units'].fillna(medias_calculadas)
```

```python
df.isnull().sum()
```

```python
# PLANO B: Afunilando o agrupamento de dados. Vamos calcular a média apenas por Marca e Segmento
```

```python
medias_marca_segmento = df.groupby(['brand', 'market_segment'])['annual_sales_units'].transform('mean')
df['annual_sales_units'] = df['annual_sales_units'].fillna(medias_marca_segmento)
```

```python
df.isnull().sum()
```
